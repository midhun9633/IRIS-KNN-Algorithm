```python
#Encoding one hot encoding
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col_transfer=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough') #based on traininng data we create a model training data   
                  # includes 3 states in testing data if any unnknown values came machune has to ignore thats y we use handle unknown
x=col_transfer.fit_transform(x)                  
```

```python
#training testing data
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.30,random_state=42)
x_test
```

```python
#creating regression model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print('constant value is',model.intercept_) #to get y intercept on x or constant
```

```python
print('slope is',model.coef_)
c=list(zip(x,model.coef_))
print(c)
```

```python
#creating df actual val and predicted val
df1=pd.DataFrame({'Actual_Value':y_test,'Predicted_value':y_pred,'difference':y_test-y_pred})
df1
```

```python
#PEM
#MAE
from sklearn.metrics import mean_absolute_error
print('MAE is',mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print("percentage",mean_absolute_percentage_error(y_test,y_pred)) #to calculate in percentage
```

```python
from sklearn.metrics import r2_score
print("r2 score is",r2_score(y_test,y_pred))
```

r2 score working: r2 score= 1-RSS/TSS
